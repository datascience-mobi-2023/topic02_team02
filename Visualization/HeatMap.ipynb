```python
#Hier Importiere ich in das Jupyter Notebook ein paar Packets
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
#Hier wird in der neuen Variable "Datensatz1" (vom Typ dataframe) folgende Datei eingelesen, das "pd." sagt uns, dass es ein befehl aus dem "pandas" package ist
Datensatz1 = pd.read_csv('/Users/dario/Documents/GitHub/topic02_team02/DMS_data/P53_HUMAN_Giacomelli_NULL_Etoposide_2018.csv')
#Jetzt zerlege ich die Werte von "mutant" da sie Angegeben sind in *AusgangsAminosäure* *Position* *NeueAminosäure* um daraufhin die Neuen Aminosäuren jeweils gegen die eine Position Plotten zu können
#Hier wird in einer neuen Spalte von "Datensatz1" namens "PositionDerMutation" die Position (also der Zahlenwert) ausgelesen und gespeichert
Datensatz1['PositionDerMutation'] = Datensatz1["mutant"].apply(lambda x: re.search(r'\d+', x).group())
#Hier wird in einer neuen Spalte von "Datensatz1" namens "Alte_AS" die ausgängliche Aminosäure ausgelesen und gespeichert
Datensatz1['Alte_AS'] = Datensatz1["mutant"].apply(lambda x: re.search(r'[A-Z]', x).group())
#Hier wird in einer neuen Spalte von "Datensatz1" namens "Neue_AS" die neue (mutierte) Aminosäure ausgelesen und gespeichert
Datensatz1['Neue_AS'] = Datensatz1["mutant"].apply(lambda x: re.search(r'[A-Z]\d+([A-Z])', x).group(1))
#In einer neuen ".csv" Datei names "DarioHeatMapTest" werden nun die ausgänglichen Daten und die "neu gewonnenen" Daten vereint
Datensatz1.to_csv('DarioHeatMapTest.csv', index=False)
```

```python
#Hier werden Als Test einmal die ersten 50 Zeilen des neuen Datensatzes "Datensatz1" mitsamt aller Spalten ausgegeben
Datensatz1.head(50)
```

```python
#Hier wird in der Variable "HeatMapTest" (vom Typ pandas.core.frame.DataFrame) erstellt, diese soll dann zu einer Heatmap werden bei der die Spalten durch "Neue_AS" aus "Datensatz1" gegeben sind und die Zeilen aus "PositionDerMutation" und "Alte_AS" gegeben sind.
#Die "Heat" der Map wird hier anhand der Werte für "DMS_score" gemessen
HeatMapTest = Datensatz1.pivot(index=['PositionDerMutation', "Alte_AS"], columns='Neue_AS', values='DMS_score')
#Hiermit erstelle ich via dem Packet "seaborn", den Daten von "HeatMapTest" und dem Farbschema "coolwarm" eine Heatmap
sns.heatmap(HeatMapTest, cmap='coolwarm')
#Sie soll den Namen "DMS Scores fpr Mutations" tragen
plt.title('DMS Scores for Mutations')
#Hiermit gebe ich die Heatmap aus
plt.show()
```
